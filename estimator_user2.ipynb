```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

```python
tsize=0.30
df_2= pd.read_csv('data_preprocessed.csv')
df_2
```

```python
X= df_2[['pm1', 'pm2', 'pm3','lum', 'temp', 'humd','pres']]
y=df_2['sm']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = tsize, random_state = 0)
rfr = RandomForestRegressor(n_estimators=100)
rfr.fit(X_train, y_train)
```

```python
df = pd.read_csv('file_name.csv') #filename of March-data from user2 in CSV format

March_features = df[['pm3','am','st', 'lum']]#taking the features
rfr_pred = rfr.predict(March_features)#generating predictions

pred_df = pd.DataFrame({'ttime':df['ttime'], 'pred_sm':rfr_pred})
pred_df.to_csv('user2_March_predictions.csv')
```
